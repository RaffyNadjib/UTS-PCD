```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Nama: ACHMAD RAFFY NADJIB
# NIM: 23424047

# 1. Baca citra grayscale
# Ganti 'nama_gambar.jpg' dengan nama file gambar yang sudah kamu upload ke Colab
img = cv2.imread('Tiger.jpg', cv2.IMREAD_GRAYSCALE)

if img is not None:
    # 2. Terapkan Rumus
    # Transformasi Negatif: G = 255 - F
    img_negatif = 255 - img

    # Transformasi Logaritmik: G = c * log(1 + F)
    # Mencari nilai konstanta c
    c = 255 / np.log(1 + np.max(img))

    # Konversi ke float32 agar perhitungan logaritma akurat, lalu kembalikan ke uint8
    img_log_float = c * np.log(1 + img.astype(np.float32))
    img_logaritmik = np.array(img_log_float, dtype=np.uint8)

    # 3. Tampilkan Keempat Gambar dan Histogram
    fig, axs = plt.subplots(3, 2, figsize=(12, 12))

    # Baris 1: Citra Asli & Histogram
    axs[0, 0].imshow(img, cmap='gray')
    axs[0, 0].set_title('Citra Asli')
    axs[0, 1].hist(img.ravel(), 256, [0,256])
    axs[0, 1].set_title('Histogram Asli')

    # Baris 2: Citra Negatif & Histogram
    axs[1, 0].imshow(img_negatif, cmap='gray')
    axs[1, 0].set_title('Citra Negatif')
    axs[1, 1].hist(img_negatif.ravel(), 256, [0,256])
    axs[1, 1].set_title('Histogram Negatif')

    # Baris 3: Citra Logaritmik & Histogram
    axs[2, 0].imshow(img_logaritmik, cmap='gray')
    axs[2, 0].set_title('Citra Logaritmik')
    axs[2, 1].hist(img_logaritmik.ravel(), 256, [0,256])
    axs[2, 1].set_title('Histogram Logaritmik')

    plt.tight_layout()
    plt.show()
else:
    print("Gambar tidak ditemukan. Pastikan file sudah di-upload ke sesi Colab.")
```
